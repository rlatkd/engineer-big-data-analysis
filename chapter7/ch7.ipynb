```python
import pandas as pd

df = pd.read_csv('./ch7.csv')
df.head(10)
```

```python
# 조건문
df['칼로리'] < 50
```

```python
# 조건에 해당하는 데이터
cond = df['칼로리'] < 50
df[cond]
df[cond].head(10)
```

```python
# 조건의 반대에 해당하는 데이터 (칼로리 50이상)
df[~cond]
```

```python
# 조건 2개 이상
cond1 = df['가격'] >= 5000
cond2 = df['칼로리'] > 100
df[cond1 & cond2] # AND
```

```python
df[cond1 | cond2] # OR
```

```python
# 문자열
cond1 = df['원산지'] == '과테말라'
cond2 = df['가격'] <= 5000
df[cond1 & cond2]
```

```python
# 문자열 포함 여부
df['메뉴'].isin(['녹차'])
```

```python
cond = df['메뉴'].isin(['녹차'])
df[cond]
```

```python
# 여러 문자열 포함 데이터
lst = ['녹차', '카푸치노', '카페라떼']
cond = df['메뉴'].isin(lst)
df[cond]
```
