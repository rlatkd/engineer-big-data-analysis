- 평가: roc-auc
- target: TravelInsurance
- 최종파일: result.csv(컬럼 1개 pred, 1확률값)

```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/3_2/train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/3_2/test.csv")

# EDA
print('\n[데이터 크기]------------------------------------------------------\n')
print(train.shape, test.shape) # test:(497, 9) -> 테스트 사이즈가 497이므로 예측해야되는 값도 497개

print('\n[데이터 샘플]------------------------------------------------------\n')
print(train.head(1))
print(test.head(1))

print('\n[데이터 정보]------------------------------------------------------\n')
print(train.info()) # object타입이 있으므로 인코딩 해야됨

print('\n[결측치 수]--------------------------------------------------------\n')
print(train.isnull().sum().sum())
print(test.isnull().sum().sum())

print('\n[타겟 빈도]--------------------------------------------------------\n')
print(train['TravelInsurance'].value_counts()) # 0:965, 1:525

# 데이터 전처리
print('\n[데이터 전처리]-----------------------------------------------------\n')
train = pd.get_dummies(train)
test = pd.get_dummies(test)
target = train.pop('TravelInsurance')
print(train.shape, test.shape, target.shape)

# 검증 데이터 분할
print('\n[검증 데이터 분할]---------------------------------------------------\n')
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)
print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)

# 머신러닝 학습
print('\n[머신러닝 학습]------------------------------------------------------\n')
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=0)
model.fit(X_tr, y_tr) # 트레이닝 피팅
# pred = model.predict_proba(X_val) # ROC, AUC일 때만 predict_proba / 나머진 predict

# 예측 및 결과 파일 생성
print('\n[예측 및 결과 파일 생성]----------------------------------------------\n')
pred = model.predict_proba(test)
result = pd.DataFrame({
	'pred': pred[:,1]
})
result.to_csv('./result.csv', index=False)

# 검토
print('\n[검토]--------------------------------------------------------------\n')
print(pd.read_csv('./result.csv').head())
print(pd.read_csv('./result.csv').shape)
```
