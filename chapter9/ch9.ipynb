```python
import pandas as pd

df = pd.read_csv('./ch9.csv')
df.head(10)
```

```python
# 값 변경 (replace)
df = df.replace('아메리카노', '룽고')
df = df.replace('녹차', '그린티')
df.head(10)
```

```python
# 값 여러개 변경 (replace)
change = {
	'룽고': '아메리카노',
	'그린티': '녹차',
}
df = df.replace(change)
df.head(10)
```

```python
# 값 변경 (loc)
df.loc[6, '원산지'] = '대한민국' # row:6/col:원산지 값 대한민국으로 변경
df.head(10)
```

```python
# 값 추가 (loc)
df.loc[1:2, '이벤트가'] = 1000 # row:1,2에 새로운 이벤트가 col로 값 1000 추가
df.head(10)
```

```python
df.to_csv('../chapter10/ch10.csv', index=False)
```

```python
import pandas as pd

df = pd.DataFrame({
	'만족도': [
		'만족',
		'보통',
		'불만',
		'보통',
		'만족'
	]
})

# map 이용
mapping = {
	'만족': 3,
	'보통': 2,
	'불만': 1
}

df['만족도_숫자'] = df['만족도'].map(mapping)

df.head(10)
```
