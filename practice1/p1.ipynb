## 문제 설명

제공된 데이터(employee_performance.csv)는 회사의 직원 정보와 근속 연수 등에 대한 정보를 포함하고 있다. 이에 수행 순서에 따라 데이터를 처리한 후, 구한 답을 소문항별로 [제출 형식]에 맞춰 답안 제출 미니페이지에 답을 입력하시오.

---

## 제공 데이터

- **직원ID**: 고유 식별자
- **부서**: 소속 부서
- **연봉**: 연간 급여
- **근속연수**: 총 근무기간
- **성과등급**: 성과 등급(A, B, C)
- **교육참여횟수**: 회사 교육에 참여한 횟수
- **고객만족도**: 담당 업무에 대한 고객의 만족도

---

## 수행 순서

1) 고객만족도가 없는 직원의 경우, 평균 고객만족도로 결측치를 제거한다.

2) 근속연수가 없는 직원의 경우, 해당 직원을 삭제한다.

3) 각 직원의 고객만족도의 성취 등 초사분위수 값을 계산한다.

4) 부서별로 평균연봉을 구하고, 두 번째로 평균연봉이 높은 부서의 평균연봉을 계산한다.

---

## 문제

① 수행순서 3)에서 계산한 값을 입력하시오.
   ⇒ **[제출 형식]** 반려하여 정수(integer)로 작성

② 수행순서 4)에서 계산한 값을 입력하시오.
   ⇒ **[제출 형식]** 반려하여 정수(integer)로 작성

```python
import pandas as pd

df = pd.read_csv("data/employee_performance.csv")

# 1)
print(df.isnull().sum()) # 결측치 개수 확인
print('---------------------------')

mean = df['고객만족도'].mean() # 평균값
df['고객만족도'] = df['고객만족도'].fillna(mean) # 결측치 채우기

print(df.isnull().sum())
print('---------------------------')

# 2)
print(df.shape) # 데이터 확인
print('---------------------------')

df = df.dropna(subset = '근속연수')

print(df.shape) # 데이터 확인
print('---------------------------')

# 3)
print(df['고객만족도'].quantile(0.75)) # 3사분위 수 값
print('---------------------------')

# 4)
print(df.groupby('부서')['연봉'].mean()) # 부서별 평균연봉
print('---------------------------')

print(df.groupby('부서')['연봉'].mean().nlargest(2)) # 부서별 평균연봉 중 큰 값 2개
print('---------------------------')
```
